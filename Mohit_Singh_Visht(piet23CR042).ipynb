```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
```

```python
df = pd.read_csv("nb_1.csv")

df.head()
```

```python
X = df.drop("Class", axis=1)
y = df["Class"]
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
y_pred_nb = nb.predict(X_test)
y_pred_proba_nb = nb.predict_proba(X_test)
```

```python
accuracy_nb = accuracy_score(y_test, y_pred_nb)

print("Model Performance:")
print(f"Accuracy: {accuracy_nb:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_nb,target_names=['Class 0', 'Class 1']))
```

```python
cm_nb = confusion_matrix(y_test, y_pred_nb)

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.heatmap(cm_nb, annot=True, fmt='d', cmap='Greens',
            xticklabels=['Class 0', 'Class 1', 'Class 2'],
            yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.title('Confusion Matrix - Naïve Bayes')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
```

```python
from sklearn.preprocessing import label_binarize

# Binarize y_test for multi-class ROC computation
# Converts classes 0,1,2 into 3 binary columns
y_test_bin = label_binarize(y_test, classes=[0,1,2])

# Number of classes
n_classes = y_test_bin.shape[1]

print("Binarized y_test shape:", y_test_bin.shape)
```

```python
from sklearn.metrics import roc_curve, auc

# Dictionaries to store FPR, TPR, and AUC for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
```

```python
# Loop through each class
# Adjust the number of classes to iterate based on the model's actual output probabilities
# as y_pred_proba_nb.shape[1] reflects the number of classes the model predicts.
num_actual_classes = y_pred_proba_nb.shape[1]

for i in range(num_actual_classes):
    # Compute False Positive Rate, True Positive Rate
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred_proba_nb[:, i])

    # Compute Area Under Curve
    roc_auc[i] = auc(fpr[i], tpr[i])

    print(f"Class {i} - AUC: {roc_auc[i]:.2f}")
```

```python
import matplotlib.pyplot as plt

colors = ['blue', 'red', 'green']

plt.figure(figsize=(6,5))

# Plot ROC curve for each class
# Use num_actual_classes as defined in the previous cell
for i, color in zip(range(num_actual_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

# Plot diagonal line (random classifier)
plt.plot([0,1],[0,1],'k--', lw=2)

plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves - Naïve Bayes')
plt.legend(loc="lower right")
plt.show()
```

```python
plt.subplot(1,3,3)

# Adjust class_names to match the actual number of classes (2)
class_names = ['Class 0', 'Class 1']
x_pos = np.arange(len(class_names))
mean_probs = np.mean(y_pred_proba_nb, axis=0)
std_probs = np.std(y_pred_proba_nb, axis=0)

plt.bar(x_pos, mean_probs, yerr=std_probs,
        align='center', alpha=0.7, ecolor='black', capsize=10)
plt.xticks(x_pos, class_names)
plt.ylabel('Average Probability')
plt.title('Prediction Probability Distribution')
plt.ylim([0, 1])

plt.tight_layout()
plt.show()
```
